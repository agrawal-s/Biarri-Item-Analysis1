```python
def Intersection(lst1, lst2): 
    return list(set(lst1).intersection(lst2))

def Union(lst1, lst2): 
    final_list = list(set(lst1) | set(lst2)) 
    return final_list 
# Driver Code 
lst1 = ['shri', 'is', 'good']
lst2 = ['shri', 'is', 'bad']
print(Intersection(lst1, lst2)) 
print(Union(lst1, lst2)) 
print(len(Intersection(lst1, lst2))/len(Union(lst1, lst2)))


def tokens(word,token_size = 3):
    return [word[i:i+token_size] for i in range(0,len(word)-token_size + 1)]
```

```python
for v1 in list1:
    for v2 in list1:
        tokens1 = tokens(v1)
        tokens2 = tokens(v2)
        if (1- len(Intersection(tokens1, tokens2))/len(Union(tokens1, tokens2))) < .7 : 
            print(v1,' , ', v2, 1 - len(Intersection(tokens1, tokens2))/len(Union(tokens1, tokens2)) )
 
```

```python
def find_jaccard(word1, word2):
    tokens1 = tokens(word1)
    tokens2 = tokens(word2)
    return (len(Intersection(tokens1, tokens2))/len(Union(tokens1, tokens2))) 
```
