```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("pokemon.csv")
```

```python
pkm = df[['attack','defense','hp','sp_attack','sp_defense','speed']]
```

```python
pkm
```

```python
corr_matrix = pkm.corr(method = 'pearson')
corr_matrix
```

```python
fig=plt.figure(figsize=(8, 8), dpi= 80, facecolor='w', edgecolor='k')
sns.heatmap(pkm.corr())
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(pkm['speed'],pkm['defense'])
```
